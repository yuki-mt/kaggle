```python
def basic_info(path: str):
    df = pd.read_csv(path)
    print('shape:', df.shape)
    print('--------------')
    print('unique numbers')
    print(df.nunique())
    print('--------------')
    print('num of duplicated:', len(df[df.duplicated()]))
    print('--------------')
    df.info()
    return df
```
